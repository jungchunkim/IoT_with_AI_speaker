## Create a new file

```
pcm.!default {
        type asym
        capture.pcm "mic"
        playback.pcm "speaker"
}

pcm.mic {
        type plug
        slave {
                pcm "hw:1,0" // "hw:<card number>,<device number>"   
        }
}

pcm.speaker{
        type plug
        slave {
                pcm "hw:0,0" // "hw:<card number>,<device number>"   
        }
}
```
