라즈베리파이에 원격 접속하기 위해 ssh 파일을 만든다.

#### 원격 접속 순서
MAC address : ether b8:27:eb:90:9a:0a"와 같이 불변하는 것 즉 기계의 고유한 번호이다.

제품의 시리얼 넘버라고 생각할 수 있다.  

세상에 단 하나밖에 없는 유니크한 값을 가지며, 변경될 수 없다. 

물리적 어뎁터이고 고유 식별 번호를 나타낸다.

###### 맥 주소는 제품 고유하지만 IP주소는 제품의 상황에 따라 계속 바뀔 수 있다.

1. WIFI 연결해서 
2. MAC주소를 이용해서 라즈베리파이의 IP주소 찾음(라즈베리 파이 IP주소) 
3. 192.168.0.52 (푸티로 원격 연결)

#### putty 로 연결하기

1. pi@(IP주소) --> IP로 입력하고 
2. raspberry --> 비밀번호


### wpa_supplicant.conf 파일

원격 접속하기 위해서 wpa_supplicant.conf 파일을 만들자.
```
ctl_interface=DIR=/var/run/wpa_supplicant
country=US
update_config=1
network={
	ssid="your_ssid"
	psk="your_pwd"
	key_mgmt=WPA-PSK
}
```
라즈베리파이를 전원코드에 연결한다.
